In this homework, Deep Q-Learning Network with episode replay and target network is implemented. Three different models are trained and compared:
1. Online training without target network effect (1 replay memory and 1 target network update step which means it is updated at every step)
2. Online training with target network (1 replay memory and N target network update steps)
3. Batch training with episode replay and target network (M replay memory and N target network update steps)

```python
import gym
import numpy as np
import matplotlib.pyplot as plt
from mlp_with_tanh import *
```

```python
env = gym.make('CartPole-v0')
env._max_episode_steps = 500
```

### 1. Online training (1 replay memory, 1 target update step)

```python
dqn1 = DQN(env, 4, 16, 2, 1, replay_memory_size=1)
dqn1.train(10000,1,1)
```

```python
dqn1.plot_scores(roll_size=100)
```

### 2. Online training with target network effect (1 replay memory, 64 target update step)

```python
dqn2 = DQN(env, 4, 16, 2, 1, replay_memory_size=1)
dqn2.train(10000,1,64)
```

```python
dqn2.plot_scores(roll_size=100)
```

### 3. Batch training (1000 replay memory, 128 batch size, 64 target update step)

```python
dqn3 = DQN(env, 4, 16, 2, 1, replay_memory_size=600)
dqn3.train(7000,128,64)
```

```python
dqn3.plot_scores(roll_size=100)
```
